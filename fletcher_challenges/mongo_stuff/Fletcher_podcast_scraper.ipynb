```python
import pandas as pd
import pickle
import feedparser
import pymongo
from pymongo import MongoClient
from time import mktime
from progress.bar import Bar
from datetime import datetime
import requests


df= pd.read_pickle('rssfeeds.pkl')
urls = df.url
```

```python
conn = pymongo.Connection()
db = conn.podcasts
db.podcast_episodes
```

```python
def parse_feeds(urls):
    print "okay, looks like we're going to parse ", len(urls), "feeds..."
    parsed = []
    bad = []
    bad_type = []
    for i, url in enumerate(urls):
        try:
            r = requests.get(url, stream=True)
            content_type = r.headers['Content-Type']
            if content_type.startswith('audio/mpeg'):
                bad.append(url)
                bad_type.append(content_type)
                print 'found some bad stuff.'
            else:
                content = r.content
                d = feedparser.parse(url)
                for entry in d['entries']:


                    ep_doc = {}
                    try:
                        ep_doc['feed_tags'] = d['feed']['tags'][0]['term']
                    except:
                        pass

                    try:
                        ep_doc['publisher'] = d['feed']['publisher']
                    except:
                        pass

                    try:
                        ep_doc['feed_description'] = d['feed']['description']
                    except:
                        pass

                    try:
                        ep_doc['feed_title'] = d['feed']['title']
                        ep_doc['episode_title'] = entry.title
                        ep_doc['episode_description'] = entry.description
                        ep_doc['published'] = entry.published
                        ep_doc['published_parsed'] = datetime.fromtimestamp(mktime(entry.published_parsed))
                        ep_doc['episode_link'] = entry.link
                        db.podcast_episodes.insert(ep_doc)
                        parsed.append(ep_doc)
                    except:
                        pass
        
        except:
            pass
#         if (i+1)%(0.1*len(urls)) == 0:
#             print "done processing feed: ", i+1, " > ", float(i+1)/float(len(urls))*100, "% done!"
#         else:
#             continue

```

```python
feed = ["http://joeroganexp.joerogan.libsynpro.com/rss", "http://cbsradionewsfeed.com/rss.php?id=92&ud=12"]
parse_feeds(feed)
```
