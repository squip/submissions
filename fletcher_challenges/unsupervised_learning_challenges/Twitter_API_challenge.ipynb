## Using Python Twitter API Wrapper - Tweepy

```python
import tweepy

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
```

```python
api = tweepy.API(auth)

irmak_tweets = api.user_timeline("essorensen", count=10)



for tweet in irmak_tweets:
    print tweet.text
    print
```

```python
query = "SF Giants"
max_tweets = 200
tweet_list = []
search_results = tweepy.Cursor(api.search, q=query).items(max_tweets)

for i, tweet in enumerate(search_results):
#     print i, tweet.id, tweet.text
    tweet_doc = {}
    tweet_doc['id'] = tweet.id
    tweet_doc['text'] = tweet.text
    tweet_doc['screen_name'] = tweet.user.screen_name
    tweet_doc['retweet_count'] = tweet.retweet_count
    tweet_doc['favorite_count'] = tweet.favorite_count
    tweet_list.append(tweet_doc)
        
tweet_list
```
