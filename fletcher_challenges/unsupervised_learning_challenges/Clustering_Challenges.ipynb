### Challenge 6

Cluster the same data with MiniBatchKMeans. MiniBatchKMeans is a fast way to apply K-means to large data without much loss -- The results are very similar. Instead of using EVERY single point to find the new place of the centroid, MiniBatch just randomly samples a small number (like 100) in the cluster to calculate the new center. Since this is usually very close to the actual center, the algorithm gets there much faster. Try it and compare the results. Example on two-feature data

```python
def minibatch_cluster_episodes(k=8, X=None):
    minibatch_model = MiniBatchKMeans(n_clusters=k)
    minibatch_clusters = minibatch_model.fit_predict(X)

    minibatch_cluster_contents = defaultdict(list)
    
    for i in range(0, len(episode_sentences)):
        minibatch_cluster_contents[minibatch_clusters[i]].append(episode_sentences[i])
        
    for i in range(0,k):    
        print "Cluster #", i+1
        pprint (minibatch_cluster_contents[i])
        print

    return (minibatch_model, minibatch_clusters, minibatch_cluster_contents)

used_k = 36        
(minibatch_model, minibatch_clusters, cluster_contents) = minibatch_cluster_episodes(used_k, X)
```
