# Challenge 1

Load the data into a pandas dataframe. 

```python
df = pd.read_csv('house-votes-84.data', header=None)
```

Replace 'y's with 1s, 'n's with 0s.

Now, almost every representative has a ?. This represents a no vote (they were absent or some other similar reason). 
If we dropped all the rows that had a ?, we would throw out most of our data. Instead, we will replace ? with the best guess in the Bayesian sense: in the absence of any other information, we will say that the probability of the representative saying YES is the ratio of others that said YES over the whole votes.

So, convert each ? to this probability (when yes=1 and no=0, this is the mean of the column)

```python
df= df.replace('y', 1)
df= df.replace('n', 0)
df= df.replace('?', np.nan)
df
```

```python
df[0]
```

```python
for column in range(1,17):
    avg = df[column].mean()
    df.loc[np.isnan(df[column]), column] = avg

df
    
```

# Challenge 2

Split the data into a test and training set. But this time, use this function: from sklearn.cross_validation import train_test_split

```python
from sklearn.cross_validation import train_test_split

X= df[range(1,17)]
Y= df[0]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=4444)
```

# Challenge 3

Using scikit.learn's KNN algorithm, train a model that predicts the party (republican/democrat) from sklearn.neighbors import KNeighborsClassifier

Try it with a lot of different k values (number of neighbors), from 1 to 20, and on the test set calculate the accuracy (number of correct predictions / number of all predictions) for each k

You can use this to calculate accuracy: from sklearn.metrics import accuracy_score

Which k value gives the highest accuracy?


```python
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
scores= []
for n in range(1,20):
    
    #fit KNN model on X,Y training values
    model = KNeighborsClassifier(n_neighbors=n).fit(X_train,Y_train)
    
    #use your model to generate Y predictions using X test data.
    Y_test_pred = model.predict(X_test)
    
    #measure accuracy of model predictions by comparing predicted Y values with Y test values.
    accuracy = accuracy_score(Y_test, Y_test_pred)
    scores.append(accuracy)
    print 'KNN (n=%i) accuracy: %.4f' % (n,accuracy)

print 'KNN max accuracy: %.4f' % max(scores)
plt.plot(range(1,20), scores)
plt.xlabel('k neighbors')
plt.ylabel('Accuracy')
```

# Challenge 4

Make a similar model but with LogisticRegression instead, calculate test accuracy.

```python
from sklearn.linear_model import LogisticRegression

#create model object that is an instance of the LogisticRegression class. fit to X, Y training data.
model = LogisticRegression().fit(X_train,Y_train)

#use your model to generate Y predictions using X test data.
Y_test_pred = model.predict(X_test)

#measure accuracy of model predictions by comparing predicted Y values with Y test values.
logit_accuracy = accuracy_score(Y_test, Y_test_pred)
print 'LogisticRegression accuracy: %.4f' % logit_accuracy
```

# Challenge 7

Plot a learning curve for the logistic regression model. But instead of going through the painstaking steps of doing it yourself, use this function:
from sklearn.learning_curve import learning_curve

This will give you the m, training errors and testing errors. All you need to do is plot them. You don't even need to give it separate training/test sets. It will do crossvalidation all by itself. Easy, isn't it? : )

```python
from sklearn.learning_curve import learning_curve
```

```python
m, train_err, test_err = learning_curve(LogisticRegression(), X,Y, train_sizes=np.linspace(0.1, 1.0, 80))
```

```python
#you get arrays with multiple train/ test error values per m. take the mean accross each row (axis 1), to keep the row count the same but the values singular. 
train_err_cv = np.mean(train_err, axis=1)
test_err_cv = np.mean(test_err, axis=1)
```

```python
#plot m values from the learning curve against the cross-validation values for train and test errors.
train, = plt.plot(m,train_err_cv)
test, = plt.plot(m,test_err_cv)

plt.legend((train,test),('training','test'),loc='best')
plt.title('Logistic Regression Learning Curve')
```

Draw the learning curve for KNN with the best k value as well.

```python
m, train_err, test_err = learning_curve(KNeighborsClassifier(n_neighbors=6), X,Y, train_sizes=np.linspace(0.1, 1.0, 80))
train_err_cv = np.mean(train_err, axis=1)
test_err_cv = np.mean(test_err, axis=1)
train, = plt.plot(m,train_err_cv)
test, = plt.plot(m,test_err_cv)

plt.legend((train,test),('training','test'),loc='best')
plt.title('KNN with k=6 Learning Curve')
```
