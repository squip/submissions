```python
%matplotlib inline
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('house-votes-84.data', header=None)
```

# Challenge 10

```python
df= df.replace('y', 1)
df= df.replace('n', 0)
df= df.replace('?', np.nan)
df
```

```python
for column in range(1,17):
    mode = df[column].mode()
    df.loc[np.isnan(df[column]), column] = mode[0]
df
```
