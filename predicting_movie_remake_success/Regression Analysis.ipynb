```python
%matplotlib inline
from collections import defaultdict
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rcParams
import requests
import pickle
import statsmodels.formula.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std

dark2_colors = [(0.10588235294117647, 0.6196078431372549, 0.4666666666666667),
                (0.8509803921568627, 0.37254901960784315, 0.00784313725490196),
                (0.4588235294117647, 0.4392156862745098, 0.7019607843137254),
                (0.9058823529411765, 0.1607843137254902, 0.5411764705882353),
                (0.4, 0.6509803921568628, 0.11764705882352941),
                (0.9019607843137255, 0.6705882352941176, 0.00784313725490196),
                (0.6509803921568628, 0.4627450980392157, 0.11372549019607843)]

rcParams['figure.figsize'] = (14, 8)
rcParams['figure.dpi'] = 150
rcParams['axes.color_cycle'] = dark2_colors
rcParams['lines.linewidth'] = 2
rcParams['axes.facecolor'] = 'white'
rcParams['font.size'] = 18
rcParams['patch.edgecolor'] = 'white'
rcParams['patch.facecolor'] = dark2_colors[0]
rcParams['font.family'] = 'normal'
rcParams['lines.markersize'] = 9
rcParams['xtick.major.pad']='8'

import prettyplotlib as ppl
```

```python
with open("original_remake_pairs.pkl", 'r') as picklefile:
    original_remake_pairs = pickle.load(picklefile)
```

```python
pd.options.display.float_format = '{:.2f}'.format
df = pd.DataFrame(original_remake_pairs)
df = df[ -np.isnan(df.rating_diff) ]
df['Win']=df['gross_difference'] > 0
df['Rating_Win']= (df['remake_rating'] - df['original_rating']) > 0
```

```python
df = pd.concat([df, pd.get_dummies(df['pair_genre'])], axis=1); df
df = pd.concat([df, pd.get_dummies(df['Rated'])], axis=1); df


df= df[(df.year_diff > 0)]
df_year_diff_10 = df[ -np.isnan(df.remake_budget) ]
df_year_diff_10['ones'] = np.ones(( len(df_year_diff_10), ))
```

### Train Linear Regression Model

```python
def train_model(dataframe, explanatory_vars, outcome_var):
    X = dataframe[explanatory_vars]
    Y = dataframe[outcome_var]
    model = sm.OLS(Y, X).fit()
    return model

# lin_model = sm.OLS(Y, X).fit()
lin_model = train_model(df_year_diff_10, ['original_box_office_inflation','original_rating','year_diff','remake_budget', 'ones'][:-1], 'remake_box_office_inflation')
lin_model.summary()
```

```python
def plot_predictions(dataframe, explanatory_vars, outcome_var, model):
    X = dataframe[explanatory_vars]
    # plot the predictions 
    ppl.plot(dataframe[outcome_var], model.predict(X), 'go')
    # plot the 1:1 actual outcome line
    ppl.plot(dataframe[outcome_var], dataframe[outcome_var], 'b')
    # axis labels
    plt.xlabel('Remake Budget (100 million)')
    plt.ylabel('Predicted Remake Gross')
```

### Build training and test set

```python
df = df_year_diff_10
N_tot = len(df)
N_train = int(round(0.76*N_tot))
N_test = N_tot - N_train

print N_train, N_test
```

```python
import random

test_rows = random.sample(df.index, N_test)
df_test = df.ix[test_rows]
df_train = df.drop(test_rows)

print len(df_train), len(df_test)
```

### Run Regression against training set

```python
training_model = train_model(df_train, ['original_box_office_inflation','original_rating','year_diff','remake_budget','ones'][:-1], 'remake_box_office_inflation')
```

```python
print training_model.summary()
```

### Run Regression against test set

```python
model_test = train_model(df_test, ['original_box_office_inflation','original_rating','year_diff','remake_budget','ones'][:-1], 'remake_box_office_inflation')
print model_test.summary()
```

### Regression Plot

```python
import matplotlib.patches as mpatches
plot_predictions(df_year_diff_10, ['original_box_office_inflation','original_rating','year_diff','remake_budget', 'ones'][:-1], 'remake_box_office_inflation', training_model)

p1 = mpatches.Rectangle((0, 0), 1, 1, fc="green")
p2 = mpatches.Rectangle((0, 0), 1, 1, fc="blue")
```

### Calculate Mean Squared Error on Training Set

```python
y_pred = training_model.predict(df_train[['original_box_office_inflation','original_rating','year_diff','remake_budget','ones'][:-1]])
```

```python
import sklearn
from sklearn.metrics import mean_squared_error
train_mse= sklearn.metrics.mean_squared_error(df_train.remake_box_office_inflation, y_pred)
train_mse
```

```python
import math
math.sqrt(train_mse)*2
```

### Calculate Mean Squared Error on Test Set

```python
y_pred = model_test.predict(df_test[['original_box_office_inflation','original_rating','year_diff','remake_budget','ones'][:-1]])
```

```python
import sklearn
from sklearn.metrics import mean_squared_error
test_mse= sklearn.metrics.mean_squared_error(df_test.remake_box_office_inflation, y_pred)
test_mse
```

```python
import math
math.sqrt(test_mse)*2
```
